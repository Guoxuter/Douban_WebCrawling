```python
# 分析爬取的数据

import pandas as pd

```

```python
with open('./data/book_data/book_reviews.csv','rb')as f:
    df=pd.read_csv(f)
print(df.shape)
df.head(-5)
```

```python
# 构造因变量
lam=17
lam1=0.01
lam2=0.021
df['y']=df[['like','dislike']].apply(lambda x:(x["like"]+lam)/(x["dislike"]+lam),axis=1)
df['y2']=df[['like','dislike']].apply(lambda x:(x["like"]+lam1)/(x["like"]+x["dislike"]+lam2),axis=1)
df.head(5)
```
